```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

def initialize_llm():
    llm = ChatGroq(
        temperature=0,
        groq_api_key="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",
        model_name="llama-3.3-70b-versatile"
    )
    return llm

def create_vector_db():
    # Ensure the 'data' directory exists and contains PDF files
    loader = DirectoryLoader("/content/data/", glob='*.pdf', loader_cls=PyPDFLoader)
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    texts = text_splitter.split_documents(documents)
    embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vector_db = Chroma.from_documents(texts, embeddings, persist_directory='./chroma_db')
    vector_db.persist()

    print("ChromaDB created and data saved")
    return vector_db

class ConversationHistory:
    def __init__(self):
        self.history = []

    def add_message(self, message_type, message):
        self.history.append((message_type, message))

    def get_context(self):
        context = "\n".join([f"{t}: {m}" for t, m in self.history])
        return context if context else "No prior conversation history."

def setup_qa_chain(vector_db, llm):
    retriever = vector_db.as_retriever()

    prompt_template = """You are a compassionate mental health chatbot. Respond thoughtfully to the following question, considering the conversation history:
    {context}
    User: {question}
    Chatbot:"""

    PROMPT = PromptTemplate(template=prompt_template, input_variables=["context", "question"])

    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs={"prompt": PROMPT}
    )
    return qa_chain

def main():
    print("Initializing Chatbot.........")
    llm = initialize_llm()

    db_path = "./chroma_db"
    if not os.path.exists(db_path):
        vector_db = create_vector_db()
    else:
        embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
        vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)

    conversation_history = ConversationHistory()  # Initialize history *outside* the loop
    qa_chain = setup_qa_chain(vector_db, llm)

    while True:
        query = input("\nHuman: ")
        if query.lower() == "exit":
            print("Chatbot: Take Care of yourself, Goodbye!")
            break

        conversation_history.add_message("Human", query)
        context = conversation_history.get_context()  # Get updated conversation history
        # Pass context and query correctly to the chain
        response = qa_chain({"query": f"{context}\nUser: {query}"})
        chatbot_reply = response['result']
        conversation_history.add_message("Chatbot", chatbot_reply)  # Save chatbot's response
        print(f"Chatbot: {chatbot_reply}")

if __name__ == "__main__":
    main()
```

```python
import datetime
import firebase_admin
from firebase_admin import credentials, firestore
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
import os
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import gradio as gr

# Firebase Connection Setup
def setup_firebase():
    if not firebase_admin._apps:
        cred = credentials.Certificate("/content/project141-22a01-firebase-adminsdk-sgnh9-c30ed8c4e9.json")
        firebase_admin.initialize_app(cred)
    return firestore.client()

# Enhanced Emotion Detection
class EmotionAnalyzer:
    def __init__(self):
        self.model_name = "j-hartmann/emotion-english-distilroberta-base"
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        self.model = AutoModelForSequenceClassification.from_pretrained(self.model_name)
        self.labels = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]

    def analyze(self, text):
        inputs = self.tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
        with torch.no_grad():
            outputs = self.model(**inputs)
        probs = torch.softmax(outputs.logits, dim=1)
        top2 = torch.topk(probs, 2)
        return {
            "primary": {
                "emotion": self.labels[top2.indices[0][0]],
                "confidence": top2.values[0][0].item()
            },
            "secondary": {
                "emotion": self.labels[top2.indices[0][1]],
                "confidence": top2.values[0][1].item()
            }
        }

# Email Service with Improved Handling
class EmailService:
    def __init__(self):
        self.sent_users = set()
        self.helpline_numbers = {
            'mumbai': 'Mpower 1on1: 1800-120-820050',
            'pune': 'Connecting NGO: +91-9922001122',
            # ... (keep all your city entries from filecode2)
        }

    def send_support_email(self, user_email, city):
        if user_email in self.sent_users:
            return "We've already sent you support resources. Please check your email."

        city = city.lower()
        sender_email = "pruthviraj.22420262@viit.ac.in"
        sender_password = "22420262"

        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = user_email
        msg['Subject'] = "Mental Health Support Resources"

        body = f"""Dear {user_email.split('@')[0]},

We genuinely care about your well-being. Here are support resources for {city.capitalize()}:
Helpline: {self.helpline_numbers.get(city, '14416 (National Mental Health Helpline)')}

Remember you're not alone. Reach out anytime.
Warm regards,
Dr. James Support Team
"""
        msg.attach(MIMEText(body, 'plain'))

        try:
            with smtplib.SMTP('smtp.gmail.com', 587) as server:
                server.starttls()
                server.login(sender_email, sender_password)
                server.sendmail(sender_email, user_email, msg.as_string())
            self.sent_users.add(user_email)
            return "Support resources sent to your email. Please check your inbox."
        except Exception as e:
            return f"Couldn't send email: {str(e)}"

# Enhanced Vector DB Management
def initialize_vector_db():
    db_path = "/content/chroma_db"
    if not os.path.exists(db_path):
        loader = DirectoryLoader("/content/data/", glob='*.pdf', loader_cls=PyPDFLoader)
        documents = loader.load()
        text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=600,
            chunk_overlap=80,
            separators=["\n\n", "\n", ". ", "! ", "? ", " ", ""]
        )
        texts = text_splitter.split_documents(documents)
        embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
        vector_db = Chroma.from_documents(texts, embeddings, persist_directory=db_path)
        vector_db.persist()
        return vector_db
    return Chroma(persist_directory=db_path,
                embedding_function=HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2'))

# Therapy Session Management
# ... (keep all the previous imports and setup code until the TherapySession class)

class TherapySession:
    def __init__(self):
        self.db = setup_firebase()
        self.email_service = EmailService()
        self.emotion_analyzer = EmotionAnalyzer()
        self.vector_db = initialize_vector_db()
        self.llm = ChatGroq(
            temperature=0.65,
            groq_api_key="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",
            model_name="llama-3.3-70b-versatile"
        )
        self.qa_chain = self.create_qa_chain()
        self.user_info = None
        self.conversation_history = []

    def create_qa_chain(self):
        prompt_template = """As Dr. James, an experienced mental health professional, conduct a therapy session with compassion:

Context: {context}
Patient: {question}

Guidelines:
1. Validate emotions first
2. Ask open-ended questions
3. Suggest practical coping strategies
4. Maintain professional boundaries
5. Escalate urgent risks appropriately

Dr. James:"""

        # Use the correct chain type and parameters
        return RetrievalQA.from_chain_type(
            llm=self.llm,
            chain_type="stuff",  # Changed from "refine" to "stuff"
            retriever=self.vector_db.as_retriever(search_kwargs={'k': 4}),
            chain_type_kwargs={
                "prompt": PromptTemplate(
                    template=prompt_template,
                    input_variables=["context", "question"]
                )
            }
        )

    def store_conversation(self, query, response, emotion_data):
        if not self.user_info:
            return

        chat_entry = {
            "query": query,
            "response": response,
            "emotions": emotion_data,
            "timestamp": datetime.datetime.now().isoformat()
        }
        self.db.collection("users").document(self.user_info['name']).collection("sessions").add(chat_entry)

    def process_message(self, message):
        # Analyze emotions
        emotion_data = self.emotion_analyzer.analyze(message)

        # Check for critical emotions
        if emotion_data['primary']['emotion'] in ['sadness', 'fear'] and emotion_data['primary']['confidence'] > 0.68:
            email_status = self.email_service.send_support_email(
                self.user_info['email'],
                self.user_info['city']
            )
        else:
            email_status = ""

        # Generate response
        context = "\n".join([f"{'Patient' if i%2==0 else 'Dr. James'}: {msg}"
                           for i, msg in enumerate(self.conversation_history[-4:])])
        result = self.qa_chain({"query": f"Context: {context}\nPatient: {message}"})

        # Update conversation
        self.conversation_history.extend([message, result['result']])
        self.store_conversation(message, result['result'], emotion_data)

        # Format output
        emotion_str = f"{emotion_data['primary']['emotion'].capitalize()} ({emotion_data['primary']['confidence']:.0%})"
        return f"Dr. James: {result['result']}\n\n[Emotion Detection: {emotion_str}] {email_status}"

# Gradio Interface
with gr.Blocks(theme=gr.themes.Soft(), title="Virtual Therapy Session") as app:
    session = gr.State(TherapySession)

    gr.Markdown("""
    # 🌸 Virtual Therapy with Dr. James
    *Confidential · Compassionate · Professional Support*
    """)

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("## Patient Profile")
            name = gr.Textbox(label="Preferred Name")
            age = gr.Number(label="Age", minimum=13, maximum=100)
            city = gr.Dropdown(["Mumbai", "Pune", "Nagpur", "Other"], label="Location")
            email = gr.Textbox(label="Email (Optional)")
            setup_btn = gr.Button("Begin Session", variant="primary")
            profile_status = gr.Markdown()

        with gr.Column(scale=2):
            gr.Markdown("## Therapy Session")
            chatbot = gr.Chatbot(height=500)
            message = gr.Textbox(label="Share your thoughts...", placeholder="Type here, press Enter to send")
            clear_btn = gr.Button("Clear History")

    def start_session(session, name, age, city, email):
        if not name or not age or not city:
            return session, "❗ Please complete all required fields (Name, Age, City)"

        session.user_info = {
            'name': name.strip(),
            'age': int(age),
            'city': city.lower(),
            'email': email or "none@provided.com"
        }
        session.db.collection("users").document(name).set(session.user_info)
        return session, f"✅ Welcome {name}! Dr. James is ready to listen."

    def chat_response(session, message, chat_history):
        if not session.user_info:
            return session, chat_history, "Please complete your profile first!"

        response = session.process_message(message)
        new_history = chat_history + [(message, response)]
        return session, new_history, ""

    setup_btn.click(
        start_session,
        [session, name, age, city, email],
        [session, profile_status]
    )

    message.submit(
        chat_response,
        [session, message, chatbot],
        [session, chatbot, message]
    )

    clear_btn.click(lambda: [], None, chatbot)

if __name__ == "__main__":
    app.launch(share=True, debug=True)
```

```python
import speech_recognition as sr
from gtts import gTTS
import os

def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print(f"Recognized: {text}")
            return text
        except sr.UnknownValueError:
            return "Sorry, I could not understand the audio."
        except sr.RequestError:
            return "Sorry, there was an issue with the speech recognition service."
```
